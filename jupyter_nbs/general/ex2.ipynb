```python
parser = PDBParser()       # create a parser
pdbs = list()

# Download all PDB structures in the previous list. If they already exist, don't download.
for id in found:
    # pypdb.get_pdb_file(id, filetype='PDB')
    pdbl.retrieve_pdb_file(id, file_format='pdb', pdir='PDB')       # Retrieve the PDB file in PDB format
    pdbs.append(parser.get_structure(str(id), 'PDB/pdb'+id+'.ent'))       # extract the structure from the file
```

```python
res_counts = list()

for pdb in pdbs:
    count = 0
    for residue in pdb.get_residues():
        count+=1
    print("PDB:", pdb, "has", count, "residues.")
    res_counts.append(count)

plt.hist(res_counts, facecolor='blue', alpha=0.5)
plt.show()
```
